```python
DATA=pd.read_csv("KaggleData.csv")
headers=DATA.values[0]
```

```python
def getting(x):
    print(DATA[DATA.keys()[x]].values[0])
    return DATA[DATA.keys()[x]].values[1:]
```

```python
TIME=getting(0)
GENDER=getting(1)
fem=GENDER=='Female'
gender=np.ones(len(GENDER))
AGES=getting(3)
gender[fem]=0
times = [int(x) for  x in TIME]



times=np.array(times)
gender=np.array(gender)

femtime=times[fem]
maltime=times[~fem]
```

```python
plt.figure(figsize=[20,5])
plt.subplot(2,1,1)
plt.hist(np.log10(maltime),bins=150)
plt.xlim(1,6)
plt.subplot(2,1,2)
plt.hist(np.log10(femtime),bins=150,color="orange")
plt.xlim(1,6)
print("")
```
