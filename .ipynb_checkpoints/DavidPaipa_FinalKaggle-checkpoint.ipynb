## Facts about Time

```python
binss=50
alp=0.65
plt.figure(figsize=[18,6])
#plt.suptitle("Duration",fontsize=FS)

timefem=(np.log10(np.array(list(map(int,getting(0))))[Gfem]))
timemal=(np.log10(np.array(list(map(int,getting(0))))[Gmal]))
timenot=(np.log10(np.array(list(map(int,getting(0))))[Gnot]))
timeself=(np.log10(np.array(list(map(int,getting(0))))[Gself]))


plt.subplot(2,1,1)
plt.ylabel("Frecuency",fontsize=fs)
plt.hist(timefem,color="b",bins=2*binss,density=True,alpha=alp,label="Female")
plt.hist(timemal,color="orange",bins=2*binss,density=True,alpha=alp,label="Male")
plt.legend()
plt.subplot(2,1,2)
plt.ylabel("Frecuency",fontsize=fs)
plt.hist(timenot,color="r",bins=binss,density=True,alpha=alp,label="Prefer not to say")
plt.hist(timeself,color="g",bins=binss,density=True,alpha=alp,label="Prefer to self-describe")
plt.legend()
plt.xlabel("$Log_{10}$ Time (in seconds)",fontsize=fs)


plt.figure(figsize=[20,15])
#plt.suptitle("About Time . . . and Genders",fontsize=FS)
plt.subplot(3,1,1)
lab=['Prefer not to say','Prefer to self-describe','Female','Male']
plt.ylabel("$Log_{10}$ Time (in seconds)",fontsize=fs)
timebp=plt.boxplot([timenot,timeself,timefem,timemal],labels=lab)


plt.subplot(3,1,2)

medians=[timebp['medians'][x].get_ydata()[0] for x in range (4)]
pos=[(timebp['medians'][x].get_xdata()[0]+timebp['medians'][x].get_xdata()[1])/2 for x in range (4)]
kurt=[stats.kurtosis(x) for x in [10**np.array(timenot),10**np.array(timeself),10**np.array(timefem),10**np.array(timemal)]]
plt.ylabel("Mean time (in seconds)",fontsize=fs)
plt.scatter(pos,10**np.array(medians),c="r",s=50)
plt.plot(pos,10**np.array(medians),linestyle="--")
plt.xticks([1,2,3,4],lab)

plt.subplot(3,1,3)
plt.scatter(pos,kurt,c="r",s=50)
plt.plot(pos,kurt,linestyle="--")
plt.ylabel("Kurtosis",fontsize=fs)
plt.xticks([1,2,3,4],lab)
print("") # evitarse la molestia
```
